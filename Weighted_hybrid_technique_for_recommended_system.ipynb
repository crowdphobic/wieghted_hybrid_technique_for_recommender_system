## Building a Basic Recommendation System

```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movie_df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movie_df.head()
```

```python
print("Credits:",credits.shape)
print("Movie Dataframe",movie_df.shape)
```

```python
credits_column_rename = credits.rename(index=str,columns={"movie_id":"id"})
movie_df_merge = movie_df.merge(credits_column_rename,on='id')
movie_df_merge.head()
```

```python
movie_cleaned_df = movie_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movie_cleaned_df.head()
```

```python
movie_cleaned_df.info()
```

## Using Weighted average for each movie's Average Rating

### W = Rv + Cm / v + m

##### where
##### W = Weighted Rating
##### R = average for the movie as a number from 0 to 10 (mean) = (rating)
##### v = number of votes for the movie = (votes)
##### m = minimum votes required to be listed in the Top 250 (currently 3000)
##### C = the mean vote across the whole report (currently 6.9)

```python
# calculate all the components based on the above formula
v = movie_cleaned_df['vote_count']
R = movie_cleaned_df['vote_average']
C = movie_cleaned_df['vote_average'].mean()
m = movie_cleaned_df['vote_count'].quantile(0.70)
```

```python
movie_cleaned_df['weighted_average'] = ((R*v) + (C*m)) / (v+m)
```

```python
movie_cleaned_df.head()
```

```python
movie_ranking = movie_cleaned_df.sort_values('weighted_average',ascending=False)
movie_ranking[['original_title','vote_average','weighted_average','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weight_average=movie_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity = movie_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax = sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),
                data=popularity)
plt.title('Most popular by Votes',weight='bold')
plt.xlabel('Score of Popularity',weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.savefig('best_populator_movies.png')
```

### Recommending based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movie_cleaned_df[['weighted_average','popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movie_cleaned_df[['normalized_weight_average','normalized_popularity']]=movie_normalized_df
```

```python
movie_cleaned_df.head()
```

```python
movie_cleaned_df['score'] = movie_cleaned_df['normalized_weight_average'] * 0.5 + movie_cleaned_df['normalized_popularity'] * 0.5
movie_scored_df = movie_cleaned_df.sort_values(['score'], ascending=False)
movie_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python

scored_df = movie_cleaned_df.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')

plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```

```python
scored_df.head(5)
```

```python
# we are recommending 2 movies(if available) from top and 2 movies(if available) from bottom on the provided movie recommendation 
def recommendation(movie):
    movie_score = scored_df[scored_df['original_title'] == movie].score.values
    recommended_movie_score_high = scored_df[scored_df['score'] > movie_score[0]].original_title.values
    recommended_movie_score_low = scored_df[scored_df['score'] < movie_score[0]].original_title.values
    recommended_movie = np.append(recommended_movie_score_high[:2], recommended_movie_score_low[:2])
    return recommended_movie    
```

```python
recommendation('Deadpool')
```
